#**ROCK** **VS** **MINE** **PREDICTION**

In this project, I've made a Machine Learning Model that predicts whether an object is a rock or a mine, based on the data that it receives from a sonar. I've used supervised ML technique Logistic Regression for that purpose. Logistic Regression is a statistical method used for binary classification and sometimes extended for multiclass classification problems. Despite its name, it's actually a classification algorithm rather than a regression algorithm.

```python
#Importing the libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

For my project, I've used the sonar dataset available on Kaggle. You can download the dataset from [here](https://www.kaggle.com/datasets/mayurdalvi/sonar-mine-dataset/download?datasetVersionNumber=1)

```python
#Loading the data
df=pd.read_csv("./sonar-mine-dataset/sonar data.csv", header=None)
```

```python
df
```

```python
df.describe()
```

Let's see how many data do we have for rock classification and how many data do we have for mine classification. M represents mine and R represents rock.

```python
df[60].value_counts()
```

Since we have almost equal number of data for each category, we don't need to filter out data further more.

```python
df.groupby(60).mean()
```

Now, we make 2 separate dataframes X and Y, where X will contain the properties to check whether the object is rock or mine, and Y will contain the outcome of whether the object is a rock or a mine.

```python
X=df.drop(columns=60,axis=1)
Y=df[60]
```

```python
X
```

```python
Y
```

Now, we can split the data into training and test data by using the train_test_split function present in the sklearn library.

```python
#Splitting the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify = Y, random_state = 1)
```

Now, we are ready to train the model by using supervised learning technique, Logistic Regression

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

Now since our model is trained, we will now check the accurecy of our model, about how accurately can it predict.

```python
#accuracy on training data
X_training_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_training_prediction, Y_train)
```

```python
training_data_accuracy
```

We can conclude that out model gives a 83% right prediction when it encounters with the data it is trained on. Now let's see its performance on test data.

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
test_data_accuracy
```

We can conclude that out model gives a 76% right prediction when it encounters with some unseen data.
